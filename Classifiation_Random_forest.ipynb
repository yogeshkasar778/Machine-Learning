 # **Random Forest Classifier Algorithm:**

**What is random forest?**

Random forest is a powerful and versatile supervised machine learning algorithm that grows and combines multiple decision trees to create a forest. It can be used for both classification and regression problems in machine learning. It is based on the concept of ensemble learning which is a processes of combining multiple classifiers to solve a complex problem and improve the performance of the model.

**Why use random forest?**

It takes less training time as campared to other algorithm. It predicts output with high accuracy, even for the large dataset it runs efficienty.It can aslo maintain accuracy when a large portion of data is missing. 

**Ensemble Method:** is a techinque that combines the predications from multiple machine learning algorithms together to make more accurate predictions than individual model. 

**Bagging:** A technique to perform ensemble decision trees. Used when our goal is to reduce the variance of a decision tree. Creates several subsets of data from the tarining sample chosen randomely with replacement. Each collection of the subset data is used to train its decision tree ending up with an ensemble of different models. 

**Bootstarp:** This process involves drawing random samples from the original dataset. Bootstarp resamples the original dataset with replacement many thousands of times to create simulated datasets.

**Python with random forest algorithm:**

```python
#loading datasets from datasets 
import pandas as pd
#importing datasets
data=pd.read_csv('iris.csv')
```

```python
data.head()
```

```python
#preprocessing of datasets
#replacing the string to number
from sklearn.preprocessing import LabelEncoder
species=LabelEncoder()
```

```python
data['species']=species.fit_transform(data['species'])
```

```python
data
```

```python
#create independant and dependanet variable
X=data[['sepal_length','sepal_width','petal_length','petal_width']]
X.head()
```

```python
y=data['species']
y
```

```python
#split datasets train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Create random forest model
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100,criterion='gini')
clf.fit(X_train,y_train)
```

```python
#predict the value
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
X_test #0:setosa, 1:versicolor, 2:virginica
```

```python
#check the score
clf.score(X_test,y_test)
```
